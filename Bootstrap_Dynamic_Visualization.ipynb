```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples from a uniform distribution between 0 and 10
samples = np.random.uniform(0, 10, size=100000)

# Plot a histogram to visualize the distribution
plt.hist(samples, bins=100, density=True)
plt.title('Histogram of Randomly Generated Samples from Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
TRUE_MEAN = samples.mean()
TRUE_MEAN
```

```python
means = []

for i in range(0,1000):
    random_samples = np.random.choice(samples, 20)
    mean_rs = random_samples.mean()
    means.append(mean_rs)

# Plot a histogram of means
plt.hist(means, bins=50, density=True)
plt.title('Histogram of Sample Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

def plot_histogram(sample_size):
    # Draw a random sample
    random_samples = np.random.choice(samples, sample_size)

    # List to store the means of the bootstrapped samples
    bootstrap_means = []

    # Number of bootstrap samples
    n_bootstrap = 10000

    for i in range(n_bootstrap):
        bootstrap_sample = np.random.choice(random_samples, len(random_samples), replace=True)
        bootstrap_means.append(bootstrap_sample.mean())

    # Plot a histogram of bootstrapped means
    plt.hist(bootstrap_means, bins=50, density=True)
    plt.title('Histogram of Bootstrapped Sample Means')
    plt.xlabel('Mean Value')
    plt.ylabel('Frequency')
    plt.show()

# Create a widget to change the sample size
sample_size_widget = widgets.IntSlider(value=50, min=3, max=200, step=10, description='Sample Size:')
widgets.interactive(plot_histogram, sample_size=sample_size_widget)
```
